```python
#1)Display SSD for the whole image. Find the location where the SSD is minimum and see if you can find Waldo there.
import numpy as np
import cv2
import matplotlib.pyplot as plt

#reading cene and both waldo templates
scene = cv2.imread('C:/Users/Filip/Google Drive/McGillSchool/Semester 7/ECSE 415/Assignments/Assignment1/where_is_waldo.jpg')
waldo = cv2.imread('C:/Users/Filip/Google Drive/McGillSchool/Semester 7/ECSE 415/Assignments/Assignment1/template1.png')
waldo_noise = cv2.imread('C:/Users/Filip/Google Drive/McGillSchool/Semester 7/ECSE 415/Assignments/Assignment1/template2.png')

#converting images format to RGB
scene = cv2.cvtColor(scene,cv2.COLOR_BGR2RGB)
waldo = cv2.cvtColor(waldo,cv2.COLOR_BGR2RGB)
waldo_noise = cv2.cvtColor(waldo_noise,cv2.COLOR_BGR2RGB)

#creating copy for displaying result
scene_disp = scene.copy()

#computing height and width of waldo template
ht, wt = waldo.shape[0:2]

#computing SSD map and finding out minimum value location
res = cv2.matchTemplate(scene, waldo, cv2.TM_SQDIFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

#printing coordinates of minimum location
print("the coordinate are " + str(min_loc))

#drawing rectangle on the scene of the detected template
top_left = min_loc
bottom_right = (top_left[0] + wt, top_left[1] + ht)
cv2.rectangle(scene_disp, top_left, bottom_right, (255, 255, 0), 5)

#dsiplaying results
plt.figure(figsize=(10,10))
plt.imshow(scene), plt.title("Input Image"), plt.xticks([]), plt.yticks([])
plt.figure(figsize=(10,10))
plt.imshow(res, cmap = 'gray'), plt.title("SSD Image"), plt.xticks([]), plt.yticks([])
plt.figure(figsize=(10,10))
plt.imshow(scene_disp), plt.title("Matched Image"), plt.xticks([]), plt.yticks([])
plt.show()
```

```python
#2)Repeat above process with the noisy template.

#copying scene for displaying result
scene_disp_2 = scene.copy()

#computing height and width of noisy waldo template
ht_2, wt_2 = waldo_noise.shape[0:2]

#computing SSD map and finding minimum value location
res_2 = cv2.matchTemplate(scene, waldo_noise, cv2.TM_SQDIFF)
min_val, max_val, min_loc_2, max_loc = cv2.minMaxLoc(res_2)

#printing location of minimum SSD value
print("the coordinate are " + str(min_loc_2))

#drawing rectangle around matched template on scene
top_left_2 = min_loc_2
bottom_right_2 = (top_left_2[0] + wt_2, top_left_2[1] + ht_2)
cv2.rectangle(scene_disp_2, top_left_2, bottom_right_2, (255, 255, 0), 5)

#displaying results
plt.figure(figsize=(10,10))
plt.imshow(scene), plt.title("Input Image"), plt.xticks([]), plt.yticks([])
plt.figure(figsize=(10,10))
plt.imshow(res_2, cmap = 'gray'), plt.title("SSD Image"), plt.xticks([]), plt.yticks([])
plt.figure(figsize=(10,10))
plt.imshow(scene_disp_2), plt.title("Matched Image"), plt.xticks([]), plt.yticks([])
plt.show()
```
